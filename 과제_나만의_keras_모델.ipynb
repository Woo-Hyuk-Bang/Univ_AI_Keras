```python
#  나만의 신경망 모델 만들기
#
# - 케라스 코드로 신경망 모델을 만들어 다음과 같은 내용으로 ppt 3-4장으로 요약 제출하시오.
#
# 1. 데이터셋 설정
# 2. 모델 설정
# 3. 최적화 및 일반화 방안 구축
# 4. 결과 및 분석
```

```python
# 1. 데이터셋 설정
# MNIST 데이터셋을 불러와서 학습용 데이터와 테스트용 데이터로 나눕니다.
# 이 데이터셋은 케라스에서 제공되므로 쉽게 불러올 수 있습니다.

from keras.datasets import mnist

# 데이터 불러오기
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# 데이터 전처리
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255

# 2. 모델 설정
# 간단한 신경망 모델을 구성하겠습니다.
# 여기서는 입력층, 은닉층, 출력층으로 이루어진
# 다층 퍼셉트론(Multi-Layer Perceptron, MLP)을 사용하겠습니다.

from keras import models
from keras import layers

# 신경망 모델 설정
model = models.Sequential()
model.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
model.add(layers.Dense(10, activation='softmax'))

# 3. 최적화 및 일반화 방안 구축
# 모델을 컴파일하고 학습시키는 과정에서
# 최적화 및 일반화를 위한 여러 가지 방법을 적용할 수 있습니다.
# 여기서는 아담 옵티마이저를 사용하여 손실 함수를 최소화하고,
# 검증 데이터를 통해 모델의 성능을 평가하겠습니다.

# 모델 컴파일
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 모델 학습
history = model.fit(train_images, train_labels, epochs=5, batch_size=128,
                    validation_split=0.2)

# 4. 결과 및 분석
# 학습이 완료된 모델의 성능을 평가하고 결과를 분석합니다.

# 모델 평가
test_loss, test_acc = model.evaluate(test_images, test_labels)
print('Test accuracy:', test_acc)
```
