```python
# keras 챕터 2 실습
# 2-1 케라스에서 MNIST 데이터셋 적재하기
from tensorflow.keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# 2-2 신경망 구조
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(512,activation="relu"),
    layers.Dense(512,activation="softmax")
])

# 2-3 컴파일 단계
model.compile(optimizer="rmsprop",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

# 2-4 이미지 데이터 준비하기
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype("float32") / 255
test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype("float32") / 255
```


2-1)
각 변수의 역할은 다음과 같습니다:

train_images: 훈련에 사용될 이미지 데이터입니다. 이 변수에는 손으로 쓴 숫자 이미지가 포함되어 있습니다. 각 이미지는 픽셀 값으로 표현되며, 훈련 데이터로 사용될 것입니다.

train_labels: train_images에 대응하는 훈련 데이터의 레이블(정답)입니다. 각 이미지에 해당하는 숫자(0부터 9까지)가 포함되어 있습니다. 이 변수는 훈련 과정에서 모델이 예측해야 할 정답을 제공합니다.

test_images: 테스트에 사용될 이미지 데이터입니다. 이 변수에도 손으로 쓴 숫자 이미지가 포함되어 있습니다. 각 이미지는 픽셀 값으로 표현되며, 테스트 데이터로 사용될 것입니다.

test_labels: test_images에 대응하는 테스트 데이터의 레이블(정답)입니다. 테스트 데이터에 대한 예측을 평가할 때 사용됩니다.

2-2)
첫 번째 Dense 레이어는 512개의 뉴런을 가지며 활성화 함수로 ReLU(Rectified Linear Activation)를 사용합니다.
두 번째 Dense 레이어는 512개의 뉴런을 가지며 활성화 함수로 소프트맥스(softmax)를 사용합니다.
이 코드는 512개의 뉴런을 가진 은닉층과 512개의 뉴런을 가진 출력층으로 구성된 간단한 신경망 모델을 만드는 것입니다.
